```python
#| context: setup
import numpy as np
import matplotlib.pyplot as plt

def create_plot(a, b):
    t = np.linspace(0, 10, 201)
    s = a * t + b
    fig, ax = plt.subplots(figsize=(8.5, 4.2))
    ax.plot(t, s, linewidth=3, color="steelblue")
    ax.axhline(0, color="black", linewidth=1)
    ax.axvline(0, color="black", linewidth=1)
    ax.set_xlim(0, 10)
    ax.set_ylim(-15, 60) # Festes Limit, damit das Bild nicht springt
    ax.set_xlabel("t")
    ax.set_ylabel("s(t)")
    ax.grid(True)
    return fig
```
